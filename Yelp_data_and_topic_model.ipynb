```python
#Prepare data for LDA
#create English stop words list
en_stop = set(stopwords.words('english'))

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()

en_stop.update((u'hotel',u'locat',u'room',u'stay',u'great',u'rooms',u'w',u'westin',u'hyatt',u'hilton',u'marriott',u'location',u'times',u'good',u'stayed',u'would',u'nice',u'square',
                u'city',u'night',u'amp',u'property',u'well',u'get',u'much',u'instead',u'us',u'time',u'year',u'could',u'<br>',u'\xc2\xa0',u'br',u'hotels',u'like',u'',u'one',
                u'whenever',u'anybody',u'also',u'really',u'tambi',u'evento',u'stk',u'make',u'de',u'que',u'el',u'este',u'para',u'mis',u'nos',u'un',u'una',u'm',
                u'visit',u'want',u'see',u'took',u'floor',u'got',u'back',u'we',u'review',u'andaz',u'taking',u'centric',u'regency',u'check',u'st'))
en_stop
```

```python
#Loop through W reviews
for i in W_set:    
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)
    # remove stop words from tokens 
    stopped_tokens = [i.decode('utf-8','ignore').encode("utf-8") for i in tokens if not i in en_stop]
    # stem tokens
    #stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    # add tokens to list
    W_text.append(stopped_tokens)                   
```

```python
#Remove the blank entries in the list
W_text2 = [[] for i in range(len(W_text))]

for i in range(len(W_text)):
    W_text2[i] = [x for x in W_text[i] if x != '']
```

```python
#Turn our tokenized documents into a id <-> term dictionary
W_dict = corpora.Dictionary(W_text2)
#Convert tokenized documents into a document-term matrix
W_corpus = [W_dict.doc2bow(text) for text in W_text2]
```

```python
#Generate LDA model
W_ldamodel = gensim.models.ldamodel.LdaModel(W_corpus, num_topics=7, id2word = W_dict, passes=20)
```

```python
print(W_ldamodel.print_topics(num_topics=7, num_words=10))
```

```python
#W hotel pyLDAvis
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(W_ldamodel, W_corpus, W_dict)
```
